```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
# data from SBB, Colorado on 29 April 2021
path = '/Users/brent/Documents/Albedo/PRISMA/20210429_BECK/albedo/PRS_20210429180418_20210429180422_0001_rdn_prj'
img = np.array(rasterio.open(path).read())
img[img==-9999] = np.NaN

# visualizing band at ~1068 nm (lighter color is snow)
# note this is in radiance
plt.imshow(img[72,:,:], cmap='Greys_r') 
plt.colorbar()
plt.title('TOA Radiance at 1068 nm')
plt.show()
```
